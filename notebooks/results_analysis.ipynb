# Results Analysis
Analyze training and deployment metrics for baseline, AMP, QAT, and TensorRT models.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

root = Path('..').resolve()
results_csv = root / 'runs' / 'benchmark_results.csv'
if results_csv.exists():
    df = pd.read_csv(results_csv)
    display(df)
else:
    print('Benchmark CSV not found yet. Run scripts/benchmark.py first.')
```

```python
if results_csv.exists():
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    df.plot(kind='bar', x='Model', y='FPS', ax=ax[0], title='FPS by Model')
    df.plot(kind='bar', x='Model', y='mAP50', ax=ax[1], title='mAP50 by Model')
    plt.tight_layout()
    plt.show()
```
